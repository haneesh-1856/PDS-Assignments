Import necessary packages

```python
import matplotlib.pyplot as plt
import pandas as pds
import seaborn as sns
```

Read the data

```python
dataset=pds.read_csv("ASGN-1 datafile.csv")
dataset
```

Data frame slicing

```python
dataset = dataset.drop(dataset.index[10:])
dataset
```

Basic Information about the dataset

```python
print("Total number of rows and columns : ",dataset.shape)
print("\nColumns Header of the Dataset : ",*list(dataset.columns,))
```

```python
#column slicing and agg function
ds = dataset.iloc[:,:4]
agg_val = ds.agg(["max", "min", "mean", "sum"])
print("\nMaximum, Minimum, Mean, and Sum values for the respective columns\n")
print(agg_val)
```

```python
from os import unlink
#Unique values in the columns
unik_val = dataset.apply(lambda i: i.unique())
print("Unique values for the columns:")
print(unik_val)
print("\nCount of unique values in each column\n",dataset.nunique())
```

```python
#Data Visualization
plt.figure(figsize=(8,6))
plt.scatter(dataset['Age '], dataset['Weight '], s=dataset['Grip strength ']*20, alpha=1.0, color='r')
plt.xlabel('Age'); plt.ylabel('Weight')
plt.title('Bubble Plot of Age vs Weight (Size by Grip Strength)')
plt.show()
```

The visualizaion shows age by weight with respect to grip strength, where grip strength is indicated with the bubble size.
