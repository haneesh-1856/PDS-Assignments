```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.csv")
df
```

**Visualizaion 1: Violin Plot of Gender by Total Scores**

This explians which race/ethnicity students scores on combined matched with respect to other race students.

```python
plt.figure(figsize=(8, 6))
sns.violinplot(y=df["math score"]+df["reading score"]+df["writing score"], x="race/ethnicity", data=df, palette="Set1")
sns.stripplot(y=df["math score"]+df["reading score"]+df["writing score"], x="race/ethnicity", data=df, color='g', alpha=0.6)
plt.title("Violin Plot of race/ethnicity by Total Scores")
plt.ylabel('Total Score')
plt.grid(True)
plt.show()
```

**Visualizaion 2: Line Plot for all scores by gender**

Explains each score wrt each gender

```python
sns.lineplot(x="gender", y="math score", data=df, label="math score", color='r')
sns.lineplot(x="gender", y="reading score", data=df, label="reading score", color='g')
sns.lineplot(x="gender", y="writing score", data=df, label="writing score", color='b')
plt.title("Line Plot of math score, reading score, and writing sccore by Group")
plt.legend(); plt.ylabel('Scores')
plt.show()
```

**Visualizaion 3: Count plot for race/ethnicity by Gender.**

Hepls to identify the different gender count in each race/ethnicity

```python
sns.countplot(x="race/ethnicity", hue="gender", data=df, palette="Set2")
plt.title("Count Plot of race/ethnicity by Gender")
plt.show()
```

**Visualization 4 : Pie Chart for different types of lunch.**



```python
plt.title('Pie chart for types of Lunch')
lunch_counts = df['lunch'].value_counts()
plt.pie(lunch_counts, labels=lunch_counts.index, autopct='%1.1f%%', colors=['b', 'r'])
plt.show()
```

**Visualization 5 : Barchart of Parental Educaion Levels**

```python
plt.figure(figsize=(11, 6))
sns.countplot(x='parental level of education', data=df,
              order=df['parental level of education'].value_counts().index)
plt.xlabel('Parental Education Level')
plt.ylabel('Count')
plt.title('Bar Chart of Parental Education Levels')
plt.show()

```
